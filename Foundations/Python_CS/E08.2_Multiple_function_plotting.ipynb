### Plot the function 

   $$
   f(x)  = \frac{\sin(x)}{x}
   $$

### from $x = -6\pi$ to $x = 6\pi$. Think carefully about which $x$ values you use when $x$ is close to zero.

```python
# Import matplotlib and numpy for plotting
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Plot inline in Jupyter notebook
%matplotlib inline
```

Create an array of values at which we want to evaluate the function: $x = -6\pi$ to $x = 6\pi$

```python
num_points = 100
x = np.linspace(-6*np.pi, 6*np.pi, num=num_points)
```

Define function:

```python
f = np.sin(x) / x
```

Plot $x$ against $f$

```python
# Plot function
plt.plot(x, f, label="$\sin(x)/x$")

# Add legend
plt.legend()

# Add axis labels
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
```

### Add to the above plot the graph of $1/ \left| x \right|$, and limit the range of the $y$ axis to 1 using `plt.ylim`. (Hint: use `np.abs(x)` to return the absolute values of each component of a NumPy array `x`. 

Define second function:

```python
g = 1 / np.abs(x)
```

Add new function to existing graph, limiting $y$ axis range to 1

```python
# Plot functions
plt.plot(x, f, label="$\sin(x)/x$")
plt.plot(x, g, label="$1/|x|$")

# Add legend
plt.legend()

# Add axis labels
plt.xlabel('$x$')
plt.ylabel('$f(x)$')

# Set y axis limits
plt.ylim(ymax = 1, ymin = -0.25);
```
