# Exercise 08.1 (function plotting)

Consider the function 

$$
f(x) = e^{x/10} \sin(\omega_{1}x)\cos(\omega_{0}x)
$$

from $x = -4\pi$ to $x = 4\pi$.  

```python
# Import matplotlib and numpy for plotting
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Import interact for interactive plotting
from ipywidgets import interact

# Plot inline in Jupyter notebook
%matplotlib inline
```

Create an array of values at which we want to evaluate the function: $x = -4\pi$ to $x = 4\pi$

```python
num_points = 100
x = np.linspace(-4*np.pi, 4*np.pi, num=num_points)
```

## 1. Plot the function when $\omega_{0} = \omega_{1} = 1$. Label the axes.

Declare $\omega_{0} = \omega_{1} = 1$

```python
ω0 = ω1 = 1
```

Define function:

```python
f = np.exp(x / 10) * np.sin(ω1 * x) * np.cos(ω0 * x)
```

Plot $x$ against $f$

```python
plt.plot(x, f)
plt.xlabel('$x$')
plt.ylabel('$e^{x/10} \sin(\omega_{1}x)\cos(\omega_{0}x)$');
```

## 2. Create an interactive plot with sliders for $\omega_{0}$ and $\omega_{1}$, varying from 0 to 2.

```python
@interact(ω0=(0, 2, 0.25), ω1=(0, 2, 0.25))
def plot(ω0=1, ω1=1):
    x = np.linspace(-4*np.pi, 4*np.pi, num=100)
    plt.xlabel('$x$')
    plt.ylabel('$e^{x/10} \sin(\omega_{1}x)\cos(\omega_{0}x)$')
    plt.title(r"$\omega_0$ = {}, $\omega_1$ = {}".format(ω0, ω1))
    plt.plot(np.exp(x / 10) * np.sin(ω1 * x) * np.cos(ω0 * x))
```
