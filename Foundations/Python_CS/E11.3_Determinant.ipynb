Determine experimentally the complexity of computing the determinant of a matrix.

```python
# Import modules for calculation and plotting
import numpy as np
import matplotlib.pyplot as plt

# Configure notebook for inline plotting
%matplotlib inline
```

Generate an $n \times n$ matrix of random numbers

```python
# Create array of problem sizes we want to test (powers of 2)
N = 2**np.arange(2, 14)

# Create matrix of random numbers
A = np.random.rand(N[-1], N[-1])
```

Time determinant computation for matrices of varying sizes

```python
# Time determinant calculation on matrices of different lengths
times = []
for n in N:
    t = %timeit -n1 -r1 -o -q np.linalg.slogdet(A[:n, :n])
    times.append(t.best)
```

Plot results of calculations to determine complexity

```python
# Plot time taken for determinant calculation
plt.loglog(N, times, marker='o', label='Determinant')

# Show reference line of O(n*log(n)) and O(n)
plt.loglog(N, 1.0e-7*N*np.log(N), label=r'$O(n \log n)$')
plt.loglog(N, 1.0e-6*N**2, label=r'$O(n^2)$')

# Add labels and legend
plt.xlabel('$n$')
plt.ylabel('$t$ (s)')
plt.legend(loc=0);
```

The determinant calculation falls between the $O(n \log n)$ and $O(n^2)$ reference lines, making it a complexity of $O(n^2)$ (worst-case scenario).
