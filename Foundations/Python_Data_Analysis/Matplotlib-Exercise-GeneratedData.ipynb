```python
# Question 1
x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Question 2
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
# Question 3
plt.plot(x, y1)
plt.plot(x, y2)
plt.axis([-2*np.pi, 2*np.pi, -1, 1])
plt.show()
```

```python
# Question 4
plt.plot(x, y1, 'r-', x, y2, 'b-', linewidth = 3.0)
plt.axis([-2*np.pi, 2*np.pi, -1, 1])
plt.show()
```

```python
# Question 5
ax = plt.subplot(111)
ax.plot(x, y1, 'r-', x, y2, 'b-', linewidth = 3.0)

ax.xaxis.set_ticks(np.linspace(-2 * np.pi, 2 * np.pi, 9))
ax.xaxis.set_ticklabels([r'$-2\pi$', r'$-\frac{3}{2}\pi$', r'$-\pi$', r'$-\frac{1}{2}\pi$', 0,
                         r'$\frac{1}{2}\pi$', r'$\pi$', r'$\frac{3}{2}\pi$', r'$2\pi$'])
ax.axis([-2*np.pi, 2*np.pi, -1, 1]);
```

```python
# Question 6
ax = plt.subplot(111)
ax.plot(x, y1, 'r-', x, y2, 'b-', linewidth = 3.0)

ax.spines['left'].set_position('zero')
ax.spines['right'].set_position('zero')
ax.spines['top'].set_position('zero')
ax.spines['bottom'].set_position('zero')

ax.xaxis.set_ticks(np.linspace(-2 * np.pi, 2 * np.pi, 9))
ax.xaxis.set_ticklabels([r'$-2\pi$', r'$-\frac{3}{2}\pi$', r'$-\pi$', r'$-\frac{1}{2}\pi$', 0,
                         r'$\frac{1}{2}\pi$', r'$\pi$', r'$\frac{3}{2}\pi$', r'$2\pi$'])
ax.axis([-2*np.pi, 2*np.pi, -1, 1]);
```

```python
# Question 7
ax = plt.subplot(111)
ax.plot(x, y1, 'r-', linewidth = 3.0, label = "sine")
ax.plot(x, y2, "b-", linewidth = 3.0, label = "cosine")

ax.spines['left'].set_position('zero')
ax.spines['right'].set_position('zero')
ax.spines['top'].set_position('zero')
ax.spines['bottom'].set_position('zero')

ax.xaxis.set_ticks(np.linspace(-2 * np.pi, 2 * np.pi, 9))
ax.xaxis.set_ticklabels([r'$-2\pi$', r'$-\frac{3}{2}\pi$', r'$-\pi$', r'$-\frac{1}{2}\pi$', 0,
                         r'$\frac{1}{2}\pi$', r'$\pi$', r'$\frac{3}{2}\pi$', r'$2\pi$'])
ax.axis([-2*np.pi, 2*np.pi, -1, 1])
ax.legend(bbox_to_anchor = (1, 0), loc = 3, borderaxespad = 0.0);
```

```python
# Question 8
y3 = np.sin(x) + np.sin(2 * x)
y4 = np.cos(x) + np.cos(2 * x)

ax = plt.subplot(111)
ax.plot(x, y1, 'r-', linewidth = 1.5, label = "sin(x)")
ax.plot(x, y2, "b-", linewidth = 1.5, label = "cos(x)")
ax.plot(x, y3, 'g:', linewidth = 1.0, label = "sin(x) + sin(2x)")
ax.plot(x, y4, "m--", linewidth = 1.0, label = "cos(x) + cos(2x)")

ax.spines['left'].set_position('zero')
ax.spines['right'].set_position('zero')
ax.spines['top'].set_position('zero')
ax.spines['bottom'].set_position('zero')

ax.xaxis.set_ticks(np.linspace(-2 * np.pi, 2 * np.pi, 9))
ax.xaxis.set_ticklabels([r'$-2\pi$', r'$-\frac{3}{2}\pi$', r'$-\pi$', r'$-\frac{1}{2}\pi$', 0,
                         r'$\frac{1}{2}\pi$', r'$\pi$', r'$\frac{3}{2}\pi$', r'$2\pi$'])
ax.axis([-2*np.pi, 2*np.pi, -2, 2])

ax.legend(bbox_to_anchor = (1, 0), loc = 3, borderaxespad = 0.0)
plt.title("Sine vs cosine patterns");
```
